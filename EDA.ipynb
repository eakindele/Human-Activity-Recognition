## Imports

```python
import pandas as pd
import numpy as np
import glob
import datetime as dt
import matplotlib.pyplot as plt
```

## Bringing in and Formatting Data

```python
# blank lists will be used to hold readins from csvs, will then combine into dataframes
sitting, standing, walking, downstairs, upstairs, jogging = [], [], [], [], [], []
```

```python
for file in glob.glob("UCI_Data/w/sitting/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'sitting'
    sitting.append(df)
sits = pd.concat(sitting)
```

```python
for file in glob.glob("UCI_Data/w/standing/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'standing'
    standing.append(df)
stands = pd.concat(standing)
```

```python
for file in glob.glob("UCI_Data/w/walk_mod/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'walking'
    walking.append(df)
walks = pd.concat(walking)
```

```python
for file in glob.glob("UCI_Data/w/jogging/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'jogging'
    jogging.append(df)
jogs = pd.concat(jogging)
```

```python
for file in glob.glob("UCI_Data/w/upstairs/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'upstairs'
    upstairs.append(df)
up_stairs = pd.concat(upstairs)
```

```python
for file in glob.glob("UCI_Data/w/downstairs/*.csv"):
    df = pd.read_csv(file)
    df['subject'] = file[-6:-4]
    df['label'] = 'downstairs'
    downstairs.append(df)
down_stairs = pd.concat(downstairs)
```

```python
df = pd.concat([sits, stands, walks, jogs, up_stairs, down_stairs])
```
